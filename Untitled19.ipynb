```python
from ngsolve import *
from netgen.geom2d import SplineGeometry
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------
# 1. Define the Original Problem (L-Shape & Exact Solution)
# ---------------------------------------------------------

def make_l_shape():
    geo = SplineGeometry()
    p1 = geo.AppendPoint(0, 0); p2 = geo.AppendPoint(1, 0)
    p3 = geo.AppendPoint(1, 1); p4 = geo.AppendPoint(-1, 1)
    p5 = geo.AppendPoint(-1, 0); p6 = geo.AppendPoint(0, -1)
    p7 = geo.AppendPoint(1, -1)
    
    geo.Append(["line", p1, p6], bc="dirichlet")
    geo.Append(["line", p6, p7], bc="dirichlet")
    geo.Append(["line", p7, p2], bc="dirichlet")
    geo.Append(["line", p2, p3], bc="dirichlet")
    geo.Append(["line", p3, p4], bc="dirichlet")
    geo.Append(["line", p4, p5], bc="dirichlet")
    geo.Append(["line", p5, p1], bc="dirichlet")
    return geo

# Define Exact Solution (Singular function)
r = sqrt(x**2 + y**2)
theta = atan2(y, x)
uex = r**(2/3) * sin(2.0/3.0 * theta)
graduex = CF((uex.Diff(x), uex.Diff(y)))

# ---------------------------------------------------------
# 2. Loop and Solve (Based on your code structure)
# ---------------------------------------------------------

errL2 = []
errH1 = []
h = []

# We use the Direct Solver method here to "Solve numerically"
for i in range(5):
    # Define mesh size h
    h_val = 0.2 / (2**i)
    h.append(h_val)
    
    # Generate Mesh
    geo = make_l_shape()
    mesh = Mesh(geo.GenerateMesh(maxh=h_val))
    
    # Setup Space
    fes = H1(mesh, order=1, dirichlet="dirichlet")
    u, v = fes.TnT()
    
    # Forms
    a = BilinearForm(fes)
    a += grad(u) * grad(v) * dx
    a.Assemble()
    
    f = LinearForm(fes)
    f.Assemble()
    
    # --- SOLVER STEP (Replacing Interpolation) ---
    gfu = GridFunction(fes)
    gfu.Set(uex, definedon=mesh.Boundaries("dirichlet"))
    
    # Homogenization (Solve for interior correction)
    r_vec = f.vec.CreateVector()
    r_vec.data = f.vec - a.mat * gfu.vec
    inv = a.mat.Inverse(freedofs=fes.FreeDofs())
    gfu.vec.data += inv * r_vec
    # ---------------------------------------------

    # Compute Error
    errL2.append(sqrt(Integrate((gfu - uex)**2, mesh)))
    errH1.append(sqrt(Integrate((grad(gfu) - graduex)**2, mesh)))

# ---------------------------------------------------------
# 3. Compute Rates & Plot (Your Code)
# ---------------------------------------------------------

def compute_rates(err_vec, nel_vec):
    rates = []
    for i in range(len(nel_vec)-1):
        r = np.log(err_vec[i] / err_vec[i+1]) / np.log(h[i] / h[i+1])
        rates.append(r)
    return rates

rates_L2 = compute_rates(errL2, h)
rates_H1 = compute_rates(errH1, h)

plt.figure(figsize=(9, 6))

# Plot errors
plt.loglog(h, errL2, marker="o", linestyle="--", label="L2 error")
plt.loglog(h, errH1, marker="s", linestyle="-.", label="H1 error")

# Annotate convergence rates
def annotate_rates(nel_vec, err_vec, rates, y_offset=0):
    for i in range(len(rates)):
        x_mid = np.sqrt(nel_vec[i] * nel_vec[i+1])
        y_mid = np.sqrt(err_vec[i] * err_vec[i+1])
        # Use white background for text to make it readable over lines
        plt.text(x_mid, y_mid*(1+y_offset), f"{rates[i]:.2f}", 
                 fontsize=9, ha="center", bbox=dict(facecolor='white', alpha=0.6, edgecolor='none'))

annotate_rates(h, errL2, rates_L2, y_offset=0.05)
annotate_rates(h, errH1, rates_H1, y_offset=0.05)

plt.xlabel("mesh size h")
plt.ylabel("Error")
plt.title("Error Convergence with Local Rates (L-Shape)")
plt.legend()
#plt.grid(True, which="both")
plt.grid(True, which="both", linestyle='--')
# Invert X axis so smaller h is on the right (optional, standard convention)
# plt.gca().invert_xaxis() 

plt.show()

print("\nFinal Calculated Rates:")
print(f"L2 Rates: {[f'{r:.2f}' for r in rates_L2]}")
print(f"H1 Rates: {[f'{r:.2f}' for r in rates_H1]}")
```
